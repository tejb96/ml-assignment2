```python
# TO DO: ADD YOUR CODE HERE FOR STEPS 3-5
# Note: for any random state parameters, you can use random_state = 0
# HINT: USING A LOOP TO STORE THE DATA IN YOUR RESULTS DATAFRAME WILL BE MORE EFFICIENT
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
#step 3
model = LogisticRegression(max_iter=2000)
model.fit(X_small, y_small)
model.fit(X,y)
model.fit(X.iloc[:,:2],y)
##step 4
result={}
datasets=[ (X, y, 'X_and_y'),(X_small, y_small, 'small data'), (X.iloc[:, :2], y, 'First_Two_Columns')]
for X_data, y_data, column in datasets:
    scores = cross_validate(model, X_data, y_data, cv=5, 
                        scoring='accuracy',
                       return_train_score=True)
    train_score=scores['train_score'].mean()
    test_score=scores['test_score'].mean()
    result[column]={'training_score':train_score,'validation_score':test_score, 'Difference':train_score-test_score}
#step 5
results=pd.DataFrame(data=result)
results.head()
```

### Step 3: Implement Machine Learning Model (1 mark)

1. Import `LinearRegression` from sklearn
2. Instantiate model `LogisticRegression(max_iter=2000)`.
3. Implement the machine learning model with `X` and `y`

```python
# TO DO: ADD YOUR CODE HERE
# Note: for any random state parameters, you can use random_state = 0
from sklearn.linear_model import LinearRegression
# from sklearn import preprocessing
# scaler = preprocessing.StandardScaler().fit(X)
# x_scaled=scaler.transform(X)

X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

### Step 4: Validate Model (1 mark)

Calculate the training and validation accuracy using mean squared error and R2 score.

```python
# TO DO: ADD YOUR CODE HERE
from sklearn.metrics import mean_squared_error, r2_score
y_tpred=lr.predict(X_train)
y_vpred=lr.predict(X_val)

r2 = r2_score(y_train, y_tpred)
mse=mean_squared_error(y_train,y_tpred)

r2_v = r2_score(y_val, y_vpred)
mse_v=mean_squared_error(y_val,y_vpred)
result={}
datasets=[ (y_train, y_tpred, 'Training accuracy'),(y_val, y_vpred, 'Validation accuracy')]
for y, y_pred, column in datasets:
    r2 = r2_score(y, y_pred)
    mse=mean_squared_error(y,y_pred)
    result[column]={'mse':mse,'r2':r2}
```

```python
# TO DO: ADD YOUR CODE HERE
from sklearn.linear_model import Ridge
from yellowbrick.datasets import load_concrete
X, y = load_concrete()
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=0)
list=[i/1000 for i in range(0,100001,100)]
for l in list:
    print(l)
    ridge10 = Ridge(alpha=l).fit(X_train, y_train)
    print("Ridge Training score: {:.2f}".format(ridge10.score(X_train, y_train)))
    print("Ridge Validation score: {:.2f}".format(ridge10.score(X_val, y_val)))
```

```python
from sklearn.linear_model import Lasso
from yellowbrick.datasets import load_concrete
X, y = load_concrete()
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=0)
list=[i for i in range(101)]
for l in list:
    print(l)
    lasso001 = Lasso(alpha=l, max_iter=100000).fit(X_train, y_train)
    print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
    print("Validation set score: {:.2f}".format(lasso001.score(X_val, y_val)))
    print("Number of features used:", np.sum(lasso001.coef_ != 0))
```
